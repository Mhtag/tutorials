```python
import requests
from bs4 import BeautifulSoup
import csv
from tabulate import *   # to prettyprint the table
```

```python
url = 'https://free-proxy-list.net/'
```

```python
html = requests.get(url)
print(html.text)
```

```python
soup = BeautifulSoup(html.text, 'lxml')
print(soup)
```

```python
table = soup.find('table')
print(table)
```

```python
rows = table.find_all('tr')
print(rows[0:2])
```

```python
header = [header.text for header in rows[0]]
print(header)
```

```python
total_row = []
for row in rows:
    single_row = [data.text for data in row if data.text != '']
    total_row.append(single_row)
    print(single_row)
       
```

```python
print(tabulate(total_row, header, tablefmt="fancy grid"))
```

```python
result = []
for row in rows:
    result.append([data.text for data in row.find_all('td')])
```

```python
print(result)
```

```python
print(tabulate(result, header, tablefmt="fancy grid"))
```

```python
# append only those value which are not zero
result = [header]
for row in rows:
    if len(row.find_all('td'))>0:
        result.append([data.text for data in row.find_all('td')])
```

```python
print(result)
```

```python
print(tabulate(result, header, tablefmt="fancy grid"))
```

```python
with open('proxies_demo.csv', 'w', newline='') as csv_file:
    writer = csv.writer(csv_file)
    writer.writerows(result)
```
