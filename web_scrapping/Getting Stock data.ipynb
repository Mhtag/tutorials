```python
import yfinance as yf
```

```python
# define the ticker symbol
tickersymbol = 'MSFT'    # Microsft
```

```python
# get the data on this ticker
tickerData = yf.Ticker(tickersymbol)
```

```python
# get the historical price for this ticker

data = tickerData.history(period='1d', start='2011-4-1', end='2021-9-28')
```

### Period : the frequenc at which to gathr the data; common option would include '1d' (daily), '1mo'(monthly), '1y'(yearly)

### start: the date to start gathering the data
### end: the date to end gathering the data.
